# Value, Wage

```python
#value column to float
def Val_Flt(x):
    if (x[-1]=='M'):
        return float(x[1:-1])*1000000
    elif(x[-1]=='K'):
        return float(x[1:-1])*1000
    else:
        return float(x[1:])
```

```python
fifa['Value']=fifa['Value'].apply(lambda x : Val_Flt(x))
```

```python
fifa['Wage']=fifa['Wage'].apply(lambda x: Val_Flt(x))
```

# Joined

```python
fifa['Joined']=pd.to_datetime(fifa['Joined'])
```

```python
fifa['Joined'] = fifa['Joined'].dt.year
```

```python
fifa['Joined'].fillna(0,inplace=True)
```

```python
fifa['Joined'] = fifa['Joined'].astype(int)
```

```python
# Contract valid until
```

```python
fifa['Contract Valid Until']=pd.to_datetime(fifa['Contract Valid Until'])
```

```python
fifa['Height'].replace(np.nan,'0',inplace=True)
```

```python
def str_flt(x):
    if x=="0":
        return (0)
    else:
        feet=int(x.split("'")[0])
        inch=int(x.split("'")[1].split("\"")[0])
        if inch<10:
            inch=inch/10
        elif 10<=inch<=12:
            inch=inch/100
            
        inchs=feet+inch
        return(float(inchs))
        
    
```

```python
fifa['Height']=fifa['Height'].apply(lambda x: str_flt(x))
```

```python
fifa['Weight'] = fifa['Weight'].str.rstrip('lbs')
```

```python
fifa['Weight'] = fifa['Weight'].astype(float)
```

```python
fifa['Release Clause'].replace(np.nan,'0',inplace=True)
```

```python
#value column to float
def rl_Flt(s):
    if (s[-1]=='M'):
        return (float(s[1:-1])*1000000)
    elif(s[-1]=='K'):
        return (float(s[1:-1])*1000)
    else:
        return (s)
    
```

```python
fifa['Release Clause']=fifa['Release Clause'].apply(lambda i : rl_Flt(i))
```

# Null

```python
# 'Overall','Value','Wage','International Reputation','Height','Weight','Release Clause'
```

```python
fifa['Weight'].fillna(fifa.Weight.mean(), inplace = True)
```

```python
fifa['Height'].fillna('5.9', inplace = True)
```

```python
wage_missing = fifa.Wage.isna()

wage_prob = fifa.Wage.value_counts(normalize=True)
fifa.loc[wage_missing, 'Wage'] = np.random.choice(wage_prob.index, p=wage_prob.values, size=wage_missing.sum())
```

```python
fifa['Contract Valid Until'].fillna(np.random.choice(fifa['Contract Valid Until']), inplace = True)
fifa['Joined'].fillna(np.random.choice(fifa['Joined']), inplace = True)
fifa['Jersey Number'].fillna(np.random.choice(fifa['Jersey Number']), inplace = True)
fifa['Club'].fillna(np.random.choice(fifa.Club), inplace = True)
fifa['Work Rate'].fillna(np.random.choice(fifa['Work Rate']), inplace = True)
fifa['International Reputation'].fillna(np.random.choice(fifa['International Reputation']), inplace = True)
```

```python
fifa.fillna(0,inplace=True)
```

```python
fifa['Overall']=fifa['Overall'].astype(float)
```

```python
fifa['Release Clause']=fifa['Release Clause'].astype(float)
```
